# Univariate Feature Selection


```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
```

```python
test = SelectKBest(score_func=chi2, k=10)
fit = test.fit(X, y)
```

```python
print(fit.scores_)
```

```python
data1= pd.DataFrame({"Column":X.columns,"Values":fit.scores_})
data1
```

```python
data1.sort_values("Values")
```

# RFE

```python
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter=400)
rfe = RFE(model, n_features_to_select=10)
fit = rfe.fit(X, y)
```

```python
fit.ranking_
```

```python
data1= pd.DataFrame({"Column":X.columns,"Values":fit.ranking_})
data1
```

```python
data1.sort_values("Values")
```

# Data Balancing

```python
Data = pd.read_csv("clean.csv")
```

```python
col =[ "InternetService","StreamingMovies","StreamingTV","gender","PhoneService"]
Data= Data.drop(col,axis=1)
```

```python
Data= Data.drop('Unnamed: 0',axis=1)
```

```python
Data
```

```python
Data["Churn"].value_counts().plot(kind="pie",autopct="%0.2f")
```

```python
#Undersampling
#Oversampling
```

```python
x = Data.drop("Churn",axis=1)
y = Data["Churn"]
```

```python
from imblearn.over_sampling import RandomOverSampler
```

```python
ros = RandomOverSampler()
X,y = ros.fit_resample(x, y)
```

```python
X
```

```python
# Save final data without encoding format
```

```python
X["Churn"] = y
```

```python
X.to_csv("Final_data.csv")
print("Done!!")
```
